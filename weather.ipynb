```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

plt.style.use('ggplot')

# 1. Data Loading and Preparation
def load_data():
    weather = pd.read_csv("us_weather.csv", index_col="DATE", parse_dates=True)
    core_weather = weather[["PRCP", "TMAX", "TMIN"]].copy()
    core_weather.columns = ["precip", "temp_max", "temp_min"]
    
    # Handle missing data
    core_weather["precip"] = core_weather["precip"].fillna(0)
    core_weather = core_weather.fillna(method="ffill").dropna()
    
    # Create targets
    core_weather["temp_target"] = core_weather["temp_max"].shift(-1)
    core_weather["precip_target"] = core_weather["precip"].shift(-1)
    
    return core_weather.dropna()

# 2. Feature Engineering
def add_features(df):
    # Lag features
    for lag in [1, 2, 3]:
        df[f"temp_max_lag{lag}"] = df["temp_max"].shift(lag)
        df[f"temp_min_lag{lag}"] = df["temp_min"].shift(lag)
        df[f"precip_lag{lag}"] = df["precip"].shift(lag)
    
    # Rolling features
    for window in [7, 14, 30]:
        df[f"temp_rolling_{window}"] = df["temp_max"].rolling(window).mean().shift(1)
        df[f"precip_rolling_{window}"] = df["precip"].rolling(window).mean().shift(1)
    
    # Seasonal features
    df["month"] = df.index.month
    df["day_of_month"] = df.index.day
    
    return df.dropna()

# 3. Model Training
def train_models(df):
    # Define predictors
    temp_predictors = ["temp_max", "temp_min", "precip"] + \
                     [col for col in df.columns if "lag" in col or "rolling" in col] + \
                     ["month", "day_of_month"]
    
    precip_predictors = ["precip", "temp_max", "temp_min"] + \
                       [col for col in df.columns if "precip_lag" in col or "precip_rolling" in col] + \
                       ["month", "day_of_month"]
    
    # Train/test split
    train = df.loc[:"2024-12-31"]
    test = df.loc["2025-01-01":]
    
    # Initialize and train models
    ridge = make_pipeline(StandardScaler(), Ridge(alpha=0.1))
    rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    
    # Temperature predictions
    ridge.fit(train[temp_predictors], train["temp_target"])
    rf.fit(train[temp_predictors], train["temp_target"])
    temp_preds = {
        "Ridge": ridge.predict(test[temp_predictors]),
        "RF": rf.predict(test[temp_predictors])
    }
    
    # Precipitation predictions
    ridge.fit(train[precip_predictors], train["precip_target"])
    rf.fit(train[precip_predictors], train["precip_target"])
    precip_preds = {
        "Ridge": ridge.predict(test[precip_predictors]),
        "RF": rf.predict(test[precip_predictors])
    }
    
    return test, temp_preds, precip_preds

# 4. Calculate Accuracy
def calculate_accuracy(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    data_range = actual.max() - actual.min()
    accuracy = 100 * (1 - mae/data_range)
    return min(100, max(0, accuracy))  # Clamp between 0-100%

# 5. Visualization Functions
def plot_temp_comparison(test, preds):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))
    
    # Calculate accuracies
    ridge_acc = calculate_accuracy(test["temp_target"], preds["Ridge"])
    rf_acc = calculate_accuracy(test["temp_target"], preds["RF"])
    
    # TMAX Plot
    ax1.plot(test.index, test["temp_target"], 'b-', label='Actual TMAX')
    ax1.plot(test.index, preds["Ridge"], 'r--', label=f'Ridge Predicted ({ridge_acc:.1f}%)')
    ax1.plot(test.index, preds["RF"], 'g:', label=f'RF Predicted ({rf_acc:.1f}%)')
    ax1.set_title("Actual vs Predicted TMAX", fontsize=14)
    ax1.set_xlabel("Date")
    ax1.set_ylabel("Temperature (°F)")
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # TMIN Plot
    ax2.plot(test.index, test["temp_min"], 'b-', label='Actual TMIN')
    ax2.set_title("Temperature Baseline", fontsize=14)
    ax2.set_xlabel("Date")
    ax2.set_ylabel("Temperature (°F)")
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def plot_precip_comparison(test, preds):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))
    
    # Calculate accuracies
    ridge_acc = calculate_accuracy(test["precip_target"], preds["Ridge"])
    rf_acc = calculate_accuracy(test["precip_target"], preds["RF"])
    
    # Precipitation Plot
    ax1.plot(test.index, test["precip_target"], 'b-', label='Actual Precipitation')
    ax1.plot(test.index, preds["Ridge"], 'r--', label=f'Ridge Predicted ({ridge_acc:.1f}%)')
    ax1.plot(test.index, preds["RF"], 'g:', label=f'RF Predicted ({rf_acc:.1f}%)')
    ax1.set_title("Actual vs Predicted Precipitation", fontsize=14)
    ax1.set_xlabel("Date")
    ax1.set_ylabel("Precipitation (inches)")
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Accuracy Heatmap by Month/Day
    results = test.copy()
    results["month"] = results.index.month
    results["day"] = results.index.day
    results["error"] = np.abs(results["precip_target"] - preds["RF"])
    results["accuracy"] = 1 - (results["error"]/results["precip_target"].max())
    
    heatmap_data = results.groupby(["month", "day"])["accuracy"].mean().unstack()
    
    sns.heatmap(heatmap_data, cmap="YlGnBu", ax=ax2)
    ax2.set_title("Precipitation Accuracy Heatmap by Month and Day", fontsize=14)
    ax2.set_xlabel("Day of Month")
    ax2.set_ylabel("Month")
    
    plt.tight_layout()
    plt.show()

# Main Execution
if __name__ == "__main__":
    weather_df = load_data()
    weather_df = add_features(weather_df)
    
    print("Training models...")
    test_data, temp_preds, precip_preds = train_models(weather_df)
    
    print("\nTemperature visualizations...")
    plot_temp_comparison(test_data, temp_preds)
    
    print("\nPrecipitation visualizations...")
    plot_precip_comparison(test_data, precip_preds)
```

```python
# Calculate accuracy for temperature predictions
temp_ridge_acc = calculate_accuracy(test_data["temp_target"], temp_preds["Ridge"])
temp_rf_acc = calculate_accuracy(test_data["temp_target"], temp_preds["RF"])

# Calculate accuracy for precipitation predictions
precip_ridge_acc = calculate_accuracy(test_data["precip_target"], precip_preds["Ridge"])
precip_rf_acc = calculate_accuracy(test_data["precip_target"], precip_preds["RF"])

# Print the accuracy results
print("\nMODEL ACCURACY RESULTS")
print("="*40)
print(f"\nTemperature Prediction Accuracy:")
print(f"Ridge Regression: {temp_ridge_acc:.1f}%")
print(f"Random Forest: {temp_rf_acc:.1f}%")

print(f"\nPrecipitation Prediction Accuracy:")
print(f"Ridge Regression: {precip_ridge_acc:.1f}%")
print(f"Random Forest: {precip_rf_acc:.1f}%")
print("="*40)
```
